```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/MatticusBa/consultinggroup/main/DATA-462%20Loqbox%20Analytics%20Project%20-%20Membership%20Downgrades%20Data%20(1).csv?token=GHSAT0AAAAAAB6DKJ2W5A53SX7P64ZONFEMY6X7Q2Q")
df

#how to i create a dataframe with only observations that have "note" in the update_reason column
# SOLUTIO
df['t/f'] = df['update_reason'].str.contains('"note"')

df = df[df['t/f'] == False]

df
```

```python
#how do i romove rows with 0 in the column activated_save, activate_spend, activated_rent and activated_coach
# SOLUTION

#df['"note" mentioned = t/f'] = df['update_reason'].str.contains('{"note":')

df = df[(df['activated_save'] != 0) | (df['activated_spend'] != 0) | (df['activated_rent'] != 0) | (df['activated_coach'] != 0)]

df
```

```python
def format_reason(update_reason):
    return(str(update_reason).replace("[",'').replace("]",'').replace("{",'').replace("}",'|').replace('"',''))

df['format_upreas'] = df['update_reason'].apply(format_reason).astype(str)

df
```

```python
df = df['update_reason'].str.split("|", expand = True)

df
```

```python
#removing rows that have a NaN in the update_reason column
df = df[df['update_reason'].notna()]

#how to extract specific text from the update_reason column
# SOLUTION
df['text_reason'] = df['update_reason'].str.extract(r'"value": "(.*?)"')

#ceating a new column that removes text from the update_reason column and replaces it with nothing
df['update_number'] = df['update_reason'].str.replace(r'\D+', '')

df['update_number'] = df['update_number'].str.replace('1', '1, ').str.replace('2', '2, ').str.replace('3', '3, ').str.replace('4', '4, ').str.replace('5', '5, ').str.replace('6', '6, ').str.replace('7', '7, ').str.replace('8', '8, ').str.replace('9', '9, ').str.replace('0', '0, ')

#removing the last comma in the update_number column
df['update_number'] = df['update_number'].str[:-2]


df
```
