```python
#extracting the numbers form the update_reason column
#creating a new column that removes text from the update_reason column and replaces it with nothing
vals['update_number'] = vals['update_reason'].str.replace(r'\D+', '')

#find "x" and replace with "x, " in update_number column
vals['update_number'] = vals['update_number'].str.replace('1', '1, ').str.replace('2', '2, ').str.replace('3', '3, ').str.replace('4', '4, ').str.replace('5', '5, ').str.replace('6', '6, ').str.replace('7', '7, ').str.replace('8', '8, ').str.replace('9', '9, ').str.replace('0', '0, ')

#removing the last comma in the update_number column
vals['update_number'] = vals['update_number'].str[:-2]

vals.head(10)
```
